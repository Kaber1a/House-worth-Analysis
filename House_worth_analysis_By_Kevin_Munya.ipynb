```python
correlation_matrix = numeric_fp.corr()
correlation_matrix
```

This shows how different variables co-rrelate. For example Store area and house price co-orelate by 82%. From this, we can conclude that Store are and house price are linearly related. If the store area increases the house price increases and vice versa.

The correlation matrix is vital as it helps in visualising a heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.show
```

This is a representation of the co-orelation matrix visually. Co-relation above 0.5 (50%) , highlights that the 2 variables strongly relate with each other and vice versa. It enables us to graphically interpret how different varibles correlate with each other and what serves as a crucial factor in decision making.

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in numeric_fp:
 plt.figure(figsize=(10, 8))
 values = numeric_fp[col]
 plt.bar(numeric_fp[col],values, color='blue')
 plt.xlabel(col)
 plt.ylabel('Values')
 plt.title(f'Distribution of {col}')
 plt.show()
```

I have used a bargraph to visually represnt distribution of the house price to the other variables(factors). All factors under analysis are linearly related to the house.

```python
import matplotlib.pyplot as plt
import seaborn as sns

for col in numeric_fp:
 plt.figure(figsize=(10, 8))
 plt.scatter(numeric_fp[col], numeric_fp['HousePrice'])
 plt.xlabel(col)
 plt.ylabel('HousePrice')
 plt.title(f'Scatter Plot of {col} vs HousePrice')
 plt.show()
```

Yes, we have indentified that there exists a linearly relationship but plotting the scatter plot of the data set enables us to identify the concentration of the data despite there being a upward sloping linear relationship
